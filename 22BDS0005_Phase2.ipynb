```python
# =============================
# 22BDS0005
# Rahel Abraham Thomas
# =============================

!pip install -q scikit-learn plotly

import io
import numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from scipy import stats

sns.set(style="whitegrid")
pd.set_option('display.max_columns', 200)

# -----------------------------
# Load dataset
# -----------------------------
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/orange_juice_withmissing.csv"
df = pd.read_csv(url)
print("Shape:", df.shape)
display(df.head())

# -----------------------------
# Clean + detect column types
# -----------------------------
df.columns = [c.strip() for c in df.columns]

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = [c for c in df.columns if c not in numeric_cols]

print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)

# -----------------------------
# Overview
# -----------------------------
print("\nMissing values per column:")
print(df.isna().sum().sort_values(ascending=False).head(20))
print("\nBasic numeric stats:")
display(df[numeric_cols].describe().T)

# -----------------------------
# 1D Analysis
# -----------------------------
for c in numeric_cols:
    series = df[c].dropna()
    if series.empty: continue
    print(f"\n== {c} ==")
    print("Skew:", series.skew(), "Kurtosis:", series.kurtosis())
    print("Quantiles:", series.quantile([0,.25,.5,.75,1.0]).to_dict())
    fig, axs = plt.subplots(1,3,figsize=(15,3))
    sns.histplot(series, kde=True, ax=axs[0]); axs[0].set_title("Hist")
    sns.boxplot(x=series, ax=axs[1]); axs[1].set_title("Box")
    sns.violinplot(x=series, ax=axs[2]); axs[2].set_title("Violin")
    plt.show()

for c in categorical_cols:
    vc = df[c].value_counts(dropna=False).head(10)
    print(f"\n== {c} == unique={df[c].nunique()}")
    display(vc)
    sns.barplot(y=vc.index.astype(str), x=vc.values)
    plt.title(f"Top {c}"); plt.show()

# -----------------------------
# 2D Analysis
# -----------------------------
if len(numeric_cols)>=2:
    corr = df[numeric_cols].corr()
    plt.figure(figsize=(8,6))
    sns.heatmap(corr, annot=True, cmap="RdYlBu", vmin=-1,vmax=1)
    plt.title("Correlation matrix"); plt.show()

if numeric_cols and categorical_cols:
    num, cat = numeric_cols[0], categorical_cols[0]
    plt.figure(figsize=(8,4))
    sns.boxplot(x=cat, y=num, data=df)
    plt.xticks(rotation=45); plt.title(f"{num} by {cat}")
    plt.show()

if len(categorical_cols)>=2:
    c1,c2 = categorical_cols[:2]
    ct = pd.crosstab(df[c1], df[c2])
    print(f"\nContingency table: {c1} vs {c2}")
    display(ct.head())
    try:
        chi2,p,_,_ = stats.chi2_contingency(ct)
        print("Chi2:",chi2,"p:",p)
    except: pass

# -----------------------------
# 3D / grouped analysis
# -----------------------------
if len(numeric_cols)>=3:
    x,y,z = numeric_cols[:3]
    sample = df[[x,y,z]].dropna().sample(min(500,len(df)))
    fig = px.scatter_3d(sample, x=x,y=y,z=z,title="3D scatter")
    fig.show()

if numeric_cols and len(categorical_cols)>=2:
    plt.figure(figsize=(10,4))
    sns.boxplot(x=categorical_cols[0], y=numeric_cols[0], hue=categorical_cols[1], data=df)
    plt.xticks(rotation=45); plt.show()

# -----------------------------
# Clustering (KMeans)
# -----------------------------
X = df[numeric_cols].dropna()
if len(X)>5 and len(numeric_cols)>0:
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)

    # elbow + silhouette
    sse, sil = [], []
    Ks = range(2,min(11,len(X)))
    for k in Ks:
        km = KMeans(n_clusters=k, random_state=42, n_init=10).fit(Xs)
        sse.append(km.inertia_)
        sil.append(silhouette_score(Xs, km.labels_))
    plt.plot(Ks,sse,'o-'); plt.title("Elbow"); plt.show()
    plt.plot(Ks,sil,'o-'); plt.title("Silhouette"); plt.show()

    best_k = Ks[int(np.argmax(sil))]
    print("Best k:",best_k)
    km = KMeans(n_clusters=best_k, random_state=42, n_init=50)
    labels = km.fit_predict(Xs)
    pca = PCA(2).fit_transform(Xs)
    pca_df = pd.DataFrame(pca, columns=['PC1','PC2']); pca_df['cluster']=labels.astype(str)
    fig = px.scatter(pca_df,x='PC1',y='PC2',color='cluster',title=f"KMeans k={best_k}")
    fig.show()
else:
    print("Not enough numeric data for clustering")
```
